```python
X=df.drop(["PCOS (Y/N)"],axis = 1)
y=df["PCOS (Y/N)"]
```

```python
from imblearn.over_sampling import RandomOverSampler
randomoversample = RandomOverSampler(random_state=888)
X_resampled, y_resampled = randomoversample.fit_resample(X,y)
y_resampled.value_counts()
```
